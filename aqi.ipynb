```python
import pandas as pd
```

```python
data = pd.read_csv('city_day.csv')
data
```

```python
data1 = data.dropna()
```

```python
data1
```

```python
data1['AQI_Bucket'].unique()
```

```python
data1['AQI_Bucket'] = data1['AQI_Bucket'].map({'Good':0,'Satisfactory':1,'Moderate':2,'Poor':3,'Very Poor':4,'Severe':5})
data1
```

```python
x = data1.drop(['City', 'Date', 'AQI_Bucket'], axis='columns').values
y = data1.AQI_Bucket.values
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
reg = RandomForestClassifier()
```

```python
reg.fit(x_train, y_train)
```

```python
reg.score(x_test, y_test)
```

```python
reg.predict(x_test)
```
